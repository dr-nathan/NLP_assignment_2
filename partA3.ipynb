```python
from simpletransformers.classification import ClassificationModel
import pandas as pd
```

```python
train = pd.read_csv("olid-train.csv", index_col = 0)
train
```

```python
train['labels'].value_counts()
```

```python
model_args = {
    "num_train_epochs": 5,
    "manual_seed": 42,
    "overwrite_output_dir": True
}

model = ClassificationModel(model_type='bert', model_name='bert-base-cased', weight=[1, 2], args=model_args) # because class 0 is only 1 third of the data 
```

```python
model.train_model(train)
```

## Model evaluation 

```python
test = pd.read_csv("olid-test.csv", index_col = 0)
test
```

```python
test['labels'].value_counts()
```

```python
result, model_outputs, wrong_predictions = model.eval_model(test)
result
```

```python
fn = result['fn']
fp = result['fp']
tn = result['tn']
tp = result['tp']
```

```python
print('Class 0:\n')

precision = tn/(tn+fn)
print(f'Precision: {precision}')

recall = tn/(tn+fp)
print(f'Recall: {recall}')

f10 = 2*((precision*recall)/(precision+recall))
print(f'F1: {f10}')
```

```python
print('Class 1:\n')

precision = tp/(tp+fp)
print(f'Precision: {precision}')

recall = tp/(tp+fn)
print(f'Recall: {recall}')

f11 = 2*((precision*recall)/(precision+recall))
print(f'F1: {f11}')
```

```python
print(f'Weighted avg F1: {(620*f10+240*f11)/860}')
print(f'Macro avg F1: {0.5*f10+0.5*f11}')
```
